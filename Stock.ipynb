```python
dt_for_corrln_APPL=dt[["Open_AAPL","High_AAPL","Low_AAPL","Adj Close_AAPL","Volume_AAPL"]]
```

```python
import seaborn as sb
```

```python
sb.heatmap(co_matrix_APPL, annot=True, cmap="coolwarm")
```

```python
import matplotlib.pyplot as plt
#import warnings
#warnings.filterwarnings("ignore")
fig, axes = plt.subplots(1,2, figsize=(10,3)) #plt.subplots()::creates a figure and one row with two columns of subplots (1 row and 2 columns).
sb.histplot(dt_for_corrln_APPL['Open_AAPL'],bins=10,ax=axes[0]) #bin::number of bins (intervals) in the histogram.
plt.boxplot(dt_for_corrln_APPL['Open_AAPL'],patch_artist = True) #patch_artist::used to fill the boxes with colors in the box plot.
axes[0].set_title('Distribution of the open price of the APPL stocks.')
axes[1].set_title('Range of the price of the APPL stocks.')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
#import warnings
#warnings.filterwarnings("ignore")
fig, axes = plt.subplots(1,2, figsize=(10,3)) #plt.subplots()::creates a figure and one row with two columns of subplots (1 row and 2 columns).
sb.histplot(dt_for_corrln_APPL['Volume_AAPL'],bins=10,ax=axes[0]) #bin::number of bins (intervals) in the histogram.
plt.boxplot(dt_for_corrln_APPL['Volume_AAPL'],patch_artist = True) #patch_artist::used to fill the boxes with colors in the box plot.
axes[0].set_title('Distribution of the volume of the APPL stocks.')
axes[1].set_title('Range of the high volume of the APPL stocks.')
plt.tight_layout()
plt.show()
```

```python
Q1=dt_for_corrln_APPL['Volume_AAPL'].quantile(0.25)

Q3=dt_for_corrln_APPL['Volume_AAPL'].quantile(0.75)

IQR= Q3 - Q1

Upperbound_vol = Q3 + 1.5*IQR


Index1=dt.loc[dt_for_corrln_APPL['Volume_AAPL']>Upperbound_vol,"Volume_AAPL"].index
#Index1:: showing the index of outliers #loc::method using for data manipulation and analysis.

dt_for_corrln_APPL=dt_for_corrln_APPL.drop(Index1)

fig, axes = plt.subplots(1,2, figsize=(10,3))
sb.histplot(dt_for_corrln_APPL['Volume_AAPL'],bins=10,ax=axes[0])
plt.boxplot(dt_for_corrln_APPL['Volume_AAPL'],patch_artist = True)
axes[0].set_title('Distribution of the volume of the APPL stocks.')
axes[1].set_title('Range of the high volume of the APPL stocks.')
plt.tight_layout()
plt.show()

```

```python
plt.scatter(dt_for_corrln_APPL['Open_AAPL'],dt_for_corrln_APPL['High_AAPL'])
plt.xlabel("Open Price")
plt.ylabel("High Price")
plt.title("Bivariate analysis of open price and high price of the stock AAPL")
plt.tight_layout()
plt.show()
```
